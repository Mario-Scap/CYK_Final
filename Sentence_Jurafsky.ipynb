```python
import import_ipynb
from nltk import CFG
from CKY import Grammar
```

```python
grammar = CFG.fromstring(
     """
S -> NP VP
S -> X1 VP
X1 -> Aux NP
S -> Verb NP
S -> X2 PP
S -> Verb PP
S -> VP PP
NP -> I | she | me
NP -> TWA | Houston
NP -> Det Nominal
NP -> Det Nominal PP
Nominal -> book | flight | meal | money | morning | elephant | pajamas | alcohol
Nominal -> Nominal Noun
Nominal -> Nominal PP
VP -> book | include | prefer | shot
VP -> Verb NP
VP -> X2 PP
X2 -> Verb NP
VP -> Verb PP
VP -> VP PP
PP -> Preposition NP
Adverb -> Adv1 Adv2
VP -> Adverb VP

Det -> that | this | the | a | an | my
Noun -> book | flight | meal | money | morning | elephant | pajamas | alcohol
Verb -> book | include | prefer | shot | drink | read
Pronoun -> I | she | me
Proper-Noun -> Houston | NWA
Aux -> does
Preposition -> from | to | on | near | through | in
Adverb -> rarely | never | always | sometimes | quickly | often
Adv1 -> rarely
Adv2 -> never
"""
)
```

```python
with open("data/Grammar_Jurasfky.txt", "w") as f:
    for rule in grammar.productions():
        f.write(str(rule) + "\n")
```

```python
sentences_list = []
sentences_list.append('book the flight through Houston ')
sentences_list.append('does she prefer a morning flight')
sentences_list.append('I shot an elephant in my pajamas')
sentences_list.append("I rarely never read that book")

for s in sentences_list:
    k = Grammar('data/Grammar_Jurasfky.txt')
    k.parse(s)
    k.print_parse_table()
    
```
